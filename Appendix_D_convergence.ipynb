# Appendix D - Unstable Adversarial Learning With Gradient Descent

Make Your First GAN With PyTorch, 2020

```python
import pandas
import matplotlib.pyplot as plt
```

```python
# set up variables, incl x y as lists for successive values
x = [0.6]
y = [0.7]

# learning rate
lr = 0.1
```

```python
# run gradient descent, adding new x y values to lists

for i in range(499):
    # dont' forget signs are different
    x.append( x[i] + lr * y[i] )
    y.append( y[i] - lr * x[i] )
```

```python
# plot successive values of x and y

df = pandas.DataFrame(data=list(zip(x,y)), columns=['x','y'])

df.plot(figsize=(20,6))
```

```python
df.plot.scatter('x','y', figsize=(10,10))
```
